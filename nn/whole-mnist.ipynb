```python
from __future__ import division
```

```python
import numpy as np
```

```python
import pandas as pd
train = pd.read_csv('../data/train.csv')
y_train = train['label'].values
del train['label']
X_train = train.values.astype(float)
```

```python
import cPickle, gzip, numpy
f = gzip.open('../data/mnist.pkl.gz', 'rb')
train_set, valid_set, test_set = cPickle.load(f)
f.close()

_X_train, _y_train = train_set[0], train_set[1]
_X_valid, _y_valid = valid_set[0], valid_set[1]
_X_test, _y_test = test_set[0], test_set[1]
```

```python
X_big = np.vstack((_X_train, _X_valid, _X_test))
y_big = np.hstack((_y_train, _y_valid, _y_test))
X_big.shape, y_big.shape
```

```python
from copper.ml.dbn.dbn import DBN
```

```python
import os
def generate_log_callback(X, y, prefix='', folder='./', report_every=1, weights_every=1, report='report.csv'):
    folder = os.path.realpath(folder)
    if not os.path.exists(folder):
        os.makedirs(folder)
    if not prefix == '':
        prefix = prefix + '-'
    report = os.path.join(folder, prefix + report)
    
    headers = ['epoch', 'iter', 'accuracy',]
    df = pd.DataFrame(columns=headers)
    df.to_csv(report, index=None)
    
    def arr2str(arr):
        np.set_printoptions(edgeitems='nan')
        s = "\"" + np.array2string(dbn.coef_, max_line_width='nan', )[1:-1].strip() + "\""
        np.set_printoptions(edgeitems='3')
        return s
        
    def cb(clf, epoch, i):
        if i % report_every == 0:
            dic = {}
            dic['epoch'] = epoch
            dic['iter'] = i
            dic['accuracy'] = (clf.predict(X) == y).mean()
            df = pd.DataFrame([dic], columns=headers)
            df.to_csv(report, index=None, header=None, mode='a')
        if i % weights_every == 0:
            np.savetxt(os.path.join(folder, prefix + 'W_e' + str(epoch) + '_i' + str(i) + '.txt'), clf.coef_)
    return cb
```

### GD

```python
cb = generate_log_callback(X_train, y_train, folder='HU25-p1-f10-GD')
```

```python
dbn = DBN([25], pretrain_epochs=0, finetune_epochs=1, progress_bars=True, finetune_callback=cb)
```

```python
dbn.fit(X_big, y_big)
```

```python
info = pd.read_csv('HU25-p1-f10-GD/report.csv')
```

```python
info['accuracy'].plot()
```
